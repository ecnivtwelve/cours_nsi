Le tri implémenté en 1ere partie renvoie la liste triée.

Nous préferions un tri en place et bien sur sans slices.

Le principe de la fonction insere en place :

image.png

L'algorithme :

+ stocker la valeur d'indice n dans une variable temporaire : val = lst[n]
+ initialiser i à n-1
+ tant que i>=0 et que lst[i]> val :
> + décale la valeur d'indice i
> réinsérer val à la position i+1

Pourquoi i+1 ? 

Il y a deux façon de sortir du while :

+ on a trouvé à l'ince i une valeur < val. Dans ce cas cette valeur n'est pas décalée et val sera insérée à la position i+1
+ on a pas trouvé de valeur < val : Dans ce cas, toutes les valeurs on été décalées, et on sort avec i = -1, on réinsère val à l'indice 0

```python
def insere(n,lst) -> list :
    """
    insere l element d'indice n à sa place
    les élément d'indice 0 à n-1 sont déjà triés
    On insère l'élément d'indice n à sa place dans la partie du tableau d'indice 0 à n-1

    """
    val = lst[n]
    print('lst = ',lst,' on va insérer à sa place :',val)
    i = n-1
    while i >= 0 and lst[i] > val :
        print(lst[i],'est plus grand que ',val,' -> on décale à droite')
        lst[i+1] = lst[i]
        i = i - 1
    lst[i+1] = val
```

```python
lst = [1,3,6,8,5,6,8]
insere(4,lst)
print(lst)
```
