**Tri par insertion récursif**

 

*   la liste vide est triée

*   pour trier la liste L, on commence par trier (récursivement) la liste L[1:],
puis on insère l'élément L[0] a dans le résultat.

Nous aurons donc besoin d'une fonction insere(x,lst) qui insère l'élément x dans une liste triée.


<font color="red" size="8"><center>Première partie : Trier une liste python</center></font>

```python
# fonction insere :
def insere(x,lst) :
  for i in range(len(lst)):
    elem = lst[i]
    if elem > x:
      n = i
      return lst[:n] + [x] + lst [n:]
      
  return lst + [x]
```

```python
# vérification
assert insere(5,[0, 2, 3, 7, 16]) == [0, 2, 3, 5, 7, 16]
assert insere(-1,[0, 2, 3, 7, 16]) == [-1 ,0, 2, 3, 7, 16]
assert insere(21,[0, 2, 3, 7, 16]) == [0, 2, 3, 7, 16,21]
assert insere('c',['a','d','r']) == ['a','c','d','r']
```

Maintenant le tri récursif lui même : 

```python
# fonction tri_insertion_recursif :
def tri_insert_recursif( lst ):
  if len(lst) == 0:
    return lst
  else:
    x = lst[0]
    return insere(x,tri_insert_recursif(lst[1:]))
```

```python
# vérification
assert tri_insert_recursif([0, 2, 3, 7, 16] )==[0, 2, 3, 7, 16]
assert tri_insert_recursif([2, 0, 3, 7, 16] )==[0, 2, 3, 7, 16]
assert tri_insert_recursif([] )==[]
assert tri_insert_recursif(['zut','bonjour','assez',"trop",'top','truc'] )==['assez','bonjour','top','trop','truc','zut']
```

<b><font color="blue">Comment fonctionne tri_insert_recursif ?</font></b>

Prenons un exemple : l = [4,2,3 1]
<pre>
1er appel  : lst=[4,2,3,1] renvoie insere [4] dans ....
2eme appel : lst=[2,3,1] renvoie insere [2] dans ....
3eme appel : lst=[3,1] renvoie insere [3] dans ....
4eme appel : lst=[1] renvoie insere [1] dans ....
5eme appel : lst=[] renvoie []
retour 4eme appel : insere [1] dans [] -> [1]
retour 3eme appel : insere [3] dans [1] -> [1,3]
retour 2eme appel : insere [2] dans [1,3] -> [1,2,3]
retour 1er appel : insere [4] dans [1,2,3] -> [1,2,3,4]
</pre>




<font color="red" size="8"><center>Deuxième partie : Trier une pile</center></font>
Nous allons utiliser l'interface d'une pile donnée ci-dessous :

```python
class Pile :
  def __init__(self):
    self.lst=[]
  def estVide(self):
    return self.lst==[]
  def empile(self,x) :
    self.lst.append(x)
  def depile(self) :
    return self.lst.pop()
  def get(self) :
    return self.lst

```

Ecrire la fonction inserePile(x,p) qui insere la valeur x
à sa place dans la pile p

Ci-dessous une version non récursive de ce problème. Nous allons le commenter et le comprendre, puis vous travaillerez sur une version récursive, bien plus élégante.
<pre>
# version non récursive : on depile p -> a
# tant que a > x et p non vide : 
#   on empile a  dans une pile temporaire
#   on depile l'élément suivant
# on doit déterminer si on est sorti de la boucle 
# parce que a &lt;x ou parce que p est vide :
# si a inférieur à x :
#   on le rempile 
#   on empile x 
#   on vide ptmp dans p
# sinon : 
#   on empile x 
#   on empile a
#   on vide ptmp dans p
</p>

```python
def inserePile(x,p) :
  if p.estVide():
    p.empile(x)
    return p
  
  a = p.depile()
  ptmp = Pile()
  
  while a > x and not p.estVide():
    ptmp.empile(a)
    a = p.depile()
    
  if a < x :
    p.empile(a)
    p.empile(x)
    while not ptmp.estVide():
      p.empile(ptmp.depile())
  else:
    p.empile(x)
    p.empile(a)
    while not ptmp.estVide():
      p.empile(ptmp.depile())
  
  return p

p=Pile()
inserePile(0,p)
inserePile(2,p)
inserePile(3,p)
inserePile(7,p)
inserePile(16,p)
inserePile(4,p)
print(p.get())
```

```python
# vérification
p=Pile()
inserePile(0,p)
inserePile(2,p)
inserePile(3,p)
inserePile(7,p)
inserePile(16,p)
assert inserePile(5,p).get() == [0, 2, 3, 5, 7, 16]
assert inserePile(-1,p).get() == [-1 ,0, 2, 3, 5, 7, 16]
assert inserePile(21,p).get() == [-1 ,0, 2, 3, 5, 7, 16, 21]
p=Pile()
p.empile('a')
p.empile('d')
p.empile('r')
assert inserePile('c',p).get() == ['a','c','d','r']
```

Version récursive de inserePile :

Adaptez l'algo précédent en version récursive
On aura pas besoin de pile temporaire ici

```python
# version récursive
def inserePile(x, p):
  if p.estVide():
    p.empile(x)
    return p
  else:
    a = p.depile()
    if a < x:
      p.empile(a)
      p.empile(x)
    else:
      inserePile(x, p)
      p.empile(a)
    return p
```

```python
# vérification
p=Pile()
inserePile(0,p)
inserePile(2,p)
inserePile(3,p)
inserePile(7,p)
inserePile(16,p)
assert inserePile(5,p).get() == [0, 2, 3, 5, 7, 16]
assert inserePile(-1,p).get() == [-1 ,0, 2, 3, 5, 7, 16]
assert inserePile(21,p).get() == [-1 ,0, 2, 3, 5, 7, 16, 21]
p=Pile()
p.empile('a')
p.empile('d')
p.empile('r')
assert inserePile('c',p).get() == ['a','c','d','r']
```

Et ré-écrire le code récursif de la fonction tri_Pile_insert_recursif( p ) pour trier la pile

```python
# fonction tri_Pile_insertion_recursif :
def tri_Pile_insert_recursif( p ):
  if p.estVide():
    return p
  else:
    x = p.depile()
    return inserePile(x, tri_Pile_insert_recursif(p))
```

```python
# vérification
p=Pile()
p.empile(10)
p.empile(2)
p.empile(32)
p.empile(7)
p.empile(16)
tri_Pile_insert_recursif( p ).get()
```
